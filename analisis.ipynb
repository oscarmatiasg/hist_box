```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv('2019.csv')

# Estadísticas descriptivas básicas
print(df.describe())

# Gráficos de líneas para mostrar las tendencias de cada variable a lo largo del tiempo
plt.figure(figsize=(10, 6))
sns.lineplot(x=df.index, y='Score', data=df)
plt.title('Tendencia del Puntaje de Felicidad')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x=df.index, y='GDP per capita', data=df)
plt.title('Tendencia del PIB per Cápita')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x=df.index, y='Healthy life expectancy', data=df)
plt.title('Tendencia de la Expectativa de Vida Saludable')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x=df.index, y='Freedom to make life choices', data=df)
plt.title('Tendencia de la Libertad para Tomar Decisiones de Vida')
plt.show()

# Gráficos de dispersión para analizar la relación entre diferentes variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Score', y='GDP per capita', data=df)
plt.title('Relación entre Puntaje de Felicidad y PIB per Cápita')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Score', y='Healthy life expectancy', data=df)
plt.title('Relación entre Puntaje de Felicidad y Expectativa de Vida Saludable')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Score', y='Freedom to make life choices', data=df)
plt.title('Relación entre Puntaje de Felicidad y Libertad para Tomar Decisiones de Vida')
plt.show()

# Combinar histogramas y boxplots para visualizar la distribución de cada variable y detectar valores atípicos
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
sns.histplot(df['Score'], ax=axs[0, 0], color='blue', edgecolor='black')
sns.boxplot(df['Score'], ax=axs[0, 0], color='blue', boxprops=dict(facecolor='blue'))
axs[0, 0].set_title('Distribución del Puntaje de Felicidad')
sns.histplot(df['GDP per capita'], ax=axs[0, 1], color='green', edgecolor='black')
sns.boxplot(df['GDP per capita'], ax=axs[0, 1], color='green', boxprops=dict(facecolor='green'))
axs[0, 1].set_title('Distribución del PIB per Cápita')
sns.histplot(df['Healthy life expectancy'], ax=axs[1, 0], color='red', edgecolor='black')
sns.boxplot(df['Healthy life expectancy'], ax=axs[1, 0], color='red', boxprops=dict(facecolor='red'))
axs[1, 0].set_title('Distribución de la Expectativa de Vida Saludable')
sns.histplot(df['Freedom to make life choices'], ax=axs[1, 1], color='yellow', edgecolor='black')
sns.boxplot(df['Freedom to make life choices'], ax=axs[1, 1], color='yellow', boxprops=dict(facecolor='yellow'))
axs[1, 1].set_title('Distribución de la Libertad para Tomar Decisiones de Vida')
plt.tight_layout()
plt.show()
```
