```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv('2019.csv')

# Explorar los datos
print(df.head())
print(df.info())
print(df.describe())

# Crear histogramas
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
sns.histplot(df['Score'], ax=axs[0, 0], color='blue', edgecolor='black')
axs[0, 0].set_title('Distribucin del Puntaje de Felicidad')
sns.histplot(df['GDP per capita'], ax=axs[0, 1], color='green', edgecolor='black')
axs[0, 1].set_title('Distribucin del PIB per Cpita')
sns.histplot(df['Healthy life expectancy'], ax=axs[1, 0], color='red', edgecolor='black')
axs[1, 0].set_title('Distribucin de la Expectativa de Vida Saludable')
sns.histplot(df['Freedom to make life choices'], ax=axs[1, 1], color='yellow', edgecolor='black')
axs[1, 1].set_title('Distribucin de la Libertad para Tomar Decisiones de Vida')
plt.tight_layout()
plt.show()

# Crear boxplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
sns.boxplot(df['Score'], ax=axs[0, 0], color='blue')
axs[0, 0].set_title('Distribucin del Puntaje de Felicidad')
sns.boxplot(df['GDP per capita'], ax=axs[0, 1], color='green')
axs[0, 1].set_title('Distribucin del PIB per Cpita')
sns.boxplot(df['Healthy life expectancy'], ax=axs[1, 0], color='red')
axs[1, 0].set_title('Distribucin de la Expectativa de Vida Saludable')
sns.boxplot(df['Freedom to make life choices'], ax=axs[1, 1], color='yellow')
axs[1, 1].set_title('Distribucin de la Libertad para Tomar Decisiones de Vida')
plt.tight_layout()
plt.show()

# Combinar histogramas y boxplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
sns.histplot(df['Score'], ax=axs[0, 0], color='blue', edgecolor='black')
sns.boxplot(df['Score'], ax=axs[0, 0], color='blue', boxprops=dict(facecolor='blue'))
axs[0, 0].set_title('Distribucin del Puntaje de Felicidad')
sns.histplot(df['GDP per capita'], ax=axs[0, 1], color='green', edgecolor='black')
sns.boxplot(df['GDP per capita'], ax=axs[0, 1], color='green', boxprops=dict(facecolor='green'))
axs[0, 1].set_title('Distribucin del PIB per Cpita')
sns.histplot(df['Healthy life expectancy'], ax=axs[1, 0], color='red', edgecolor='black')
sns.boxplot(df['Healthy life expectancy'], ax=axs[1, 0], color='red', boxprops=dict(facecolor='red'))
axs[1, 0].set_title('Distribucin de la Expectativa de Vida Saludable')
sns.histplot(df['Freedom to make life choices'], ax=axs[1, 1], color='yellow', edgecolor='black')
sns.boxplot(df['Freedom to make life choices'], ax=axs[1, 1], color='yellow', boxprops=dict(facecolor='yellow'))
axs[1, 1].set_title('Distribucin de la Libertad para Tomar Decisiones de Vida')
plt.tight_layout()
plt.show()
```
